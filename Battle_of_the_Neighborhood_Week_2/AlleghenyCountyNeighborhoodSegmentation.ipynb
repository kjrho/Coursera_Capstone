### School District Ranking
I used schooldigger.com's API to pull the school rankings for Allegheny County. I grouped the school rankings into 5 tiers

```python
sdf = pd.read_csv('schooldigger_ranking.csv'
                    , usecols=('Rank','DistrictName','County'))
srank = sdf[(sdf['County'] =='Allegheny')]

srank.sort_values(by=['Rank'])
srank.head()

#Create school tier classification
def school_class(row):
    if row["Rank"] <= 7 :
        return "1"
    elif row["Rank"] >=8 and (row["Rank"] < 15):
        return "2"
    elif row["Rank"] >=15 and (row["Rank"] < 25):
        return "3"
    elif row["Rank"] >=26 and (row["Rank"] < 30):
        return "4"
    else:
        return "5"

srank["SCHOOL_TIER"] = srank.apply(school_class, axis=1)


rnk = srank.sort_values(by=['Rank', 'SCHOOL_TIER']).reset_index()
rnk.columns = ['index','Rank','SCHOOLDESC','County','SCHOOL_TIER']
rnk.head()
```

```python
df1 = df.merge(rnk,on='SCHOOLDESC',how="left")

df1.drop(["index","County"],inplace=True,axis=1)
df1
```

```python
u = df1[['Rank']]
df1[['Rank']] = u.fillna(50)
```

### Real Estate Sales 2018
I will use the assessment file to get the average home sales for 2018 by census tract

```python
recent_sls = parcels[
                  (parcels['SALEDESC'].isin(['VALID SALE','OTHER VALID']) &
                  (parcels['SALEYEAR']==2018) &
                  (parcels['CLASSDESC']=='RESIDENTIAL') &
                  (parcels['SALEPRICE']<=800000) &
                  (parcels['LOTAREA']<=15000)&
                  (parcels['USEDESC'].isin(['SINGLE FAMILY','APART: 5-19 UNITS','CONDOMINIUM','FOUR FAMILY','ROWHOUSE','THREE FAMILY','TOWNHOUSE','TWO FAMILY'])) 
                  )].copy()
recent_sls.describe()
```

```python
rd = recent_sls[['PARID','geo_id_tract','Group_Neighborhoods','MUNICODE','MUNIDESC','SCHOOLCODE','SCHOOLDESC','USECODE',
            'USEDESC','YEARBLT','FINISHEDLIVINGAREA','FULLBATHS','BEDROOMS','SALEPRICE','LOTAREA','SALEYEAR','SCHOOL_TIER']].copy()
rd.columns
```

```python
rd.head()
```

```python
rd['SLSPRICE_FOR_SQ'] = np.where(rd['FINISHEDLIVINGAREA']>0,rd['SALEPRICE'],0)
rd['SLSPRICE_FOR_LOT'] = np.where(rd['LOTAREA']>0,rd['SALEPRICE'],0)

rd['PRICEPERSQFT'] = round(rd['SLSPRICE_FOR_SQ']/rd['FINISHEDLIVINGAREA'],0)

rd['PARID_CNT'] = 1    
```

```python
def classtype(row):
    if row['USEDESC'] == 'SINGLE FAMILY':
        return 'SINGLE_FAMILY'
    elif row['USEDESC'] =='APART: 5-19 UNITS':
        return 'APT_CONDO' 
    elif row['USEDESC'] =='CONDOMINIUM':
        return 'APT_CONDO' 
    elif row['USEDESC'] =='APART: 5-19 UNITS':
        return 'APT_CONDO' 
    elif row['USEDESC'] =='FOUR FAMILY':
        return 'MULTI_FAMILY' 
    elif row['USEDESC'] =='ROWHOUSE':
        return 'MULTI_FAMILY' 
    elif row['USEDESC'] =='THREE FAMILY':
        return 'MULTI_FAMILY' 
    elif row['USEDESC'] =='TOWNHOUSE':
        return 'MULTI_FAMILY' 
    elif row['USEDESC'] =='TWO FAMILY':
        return 'MULTI_FAMILY' 
    else:
        return 'OTHER'

rd['CLASS_TYPE'] = rd.apply(classtype, axis=1)
```

```python
rd.head()
```

```python
rd['SLSPRICE_APT_CONDO'] = np.where(rd['CLASS_TYPE']=='APT_CONDO',rd['SALEPRICE'],0)
rd['SLSPRICE_SINGLE_FAMILY'] = np.where(rd['CLASS_TYPE']=='SINGLE_FAMILY',rd['SALEPRICE'],0)
rd['SLSPRICE_MULTI'] = np.where(rd['CLASS_TYPE']=='MULTI_FAMILY',rd['SALEPRICE'],0)

rd['LOT_APT_CONDO'] = np.where(rd['CLASS_TYPE']=='APT_CONDO',rd['LOTAREA'],0)
rd['LOT_SINGLE_FAMILY'] = np.where(rd['CLASS_TYPE']=='SINGLE_FAMILY',rd['LOTAREA'],0)
rd['LOT_MULTI'] = np.where(rd['CLASS_TYPE']=='MULTI_FAMILY',rd['LOTAREA'],0)

rd['CNT_APT_CONDO'] = np.where(rd['CLASS_TYPE']=='APT_CONDO',rd['PARID_CNT'] ,0)
rd['CNT_SINGLE_FAMILY'] = np.where(rd['CLASS_TYPE']=='SINGLE_FAMILY',rd['PARID_CNT'] ,0)
rd['CNT_MULTI'] = np.where(rd['CLASS_TYPE']=='MULTI_FAMILY',rd['PARID_CNT'] ,0)

rd['PSQ_APT_CONDO'] = np.where(rd['CLASS_TYPE']=='APT_CONDO',round(rd['SLSPRICE_FOR_SQ']/rd['FINISHEDLIVINGAREA'],0),0)
rd['PSQ_SINGLE_FAMILY'] = np.where(rd['CLASS_TYPE']=='SINGLE_FAMILY',round(rd['SLSPRICE_FOR_SQ']/rd['FINISHEDLIVINGAREA'],0),0)
rd['PSQ_MULTI'] = np.where(rd['CLASS_TYPE']=='MULTI_FAMILY',round(rd['SLSPRICE_FOR_SQ']/rd['FINISHEDLIVINGAREA'],0),0)

rd['PRICEPERSQFT'] = round(rd['SLSPRICE_FOR_SQ']/rd['FINISHEDLIVINGAREA'],0)

rd['HOME_AGE'] = round(2019-rd['YEARBLT'],0)
```

```python
#Pivot by census tract and part of town
ct_grouped = pd.pivot_table(rd, values=['SALEPRICE', 'LOTAREA','PARID_CNT','HOME_AGE','PRICEPERSQFT','SLSPRICE_APT_CONDO'
                                  ,'SLSPRICE_SINGLE_FAMILY','SLSPRICE_MULTI','LOT_APT_CONDO'
                                  ,'LOT_SINGLE_FAMILY','LOT_MULTI','CNT_APT_CONDO','CNT_SINGLE_FAMILY'
                                  ,'CNT_MULTI','PSQ_APT_CONDO','PSQ_SINGLE_FAMILY'
                                  ,'PSQ_MULTI','FULLBATHS','BEDROOMS']
                       , index=['geo_id_tract'],
                       aggfunc={'SALEPRICE': np.mean,'SALEPRICE': np.sum,'PARID_CNT':np.sum, 'HOME_AGE':np.mean,'PRICEPERSQFT':np.mean,
                                'LOTAREA': np.mean,'SLSPRICE_APT_CONDO': np.mean
                                ,'SLSPRICE_SINGLE_FAMILY': np.mean,'SLSPRICE_MULTI': np.mean,'LOT_APT_CONDO': np.mean
                                ,'LOT_SINGLE_FAMILY': np.mean,'LOT_MULTI': np.mean,'CNT_APT_CONDO': np.sum,'CNT_SINGLE_FAMILY': np.sum,
                                'CNT_MULTI': np.sum,'PSQ_APT_CONDO': np.mean,'PSQ_SINGLE_FAMILY': np.mean
                                ,'PSQ_MULTI': np.mean, 'FULLBATHS':np.mean, 'BEDROOMS':np.mean
                             })
ct_grouped = ct_grouped.reset_index()

ct_grouped.rename(index=str, columns={'geo_id_tract':'GEO_ID','SALEPRICE':'AVGSALEPRICE','LOTAREA':'AVGLOTAREA','PARID_CNT':'PARCELCOUNT'},inplace = True) 
ct_grouped.head()
```

### Exploratory Data Analysis

```python
#Classify Neighborhoods
# one hot encoding
ac_onehot = pd.get_dummies(parcels[['USEDESC']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
ac_onehot['GEO_ID']  = parcels['geo_id_tract']

# move neighborhood column to the first column
fixed_columns = [ac_onehot.columns[-1]] + list(ac_onehot.columns[:-1])
ac_onehot = ac_onehot[fixed_columns]

ac_onehot.head()
```

```python
ac_grouped = ac_onehot.groupby('GEO_ID').mean().reset_index()
ac_grouped.head()
```

```python
#Use the class code to determine what types of properties are in each census tract
num_top_props = 5

for hood in ac_grouped['GEO_ID']:
    #print("----"+hood+"----")
    temp = ac_grouped[ac_grouped['GEO_ID'] == hood].T.reset_index()
    temp.columns = ['prop','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_props))
    print('\n')
```

```python
def return_most_common_props(row, num_top_props):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_props]
```

```python
num_top_props = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top properties
columns = ['GEO_ID']
for ind in np.arange(num_top_props):
    try:
        columns.append('{}{} Most Common Property'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Property'.format(ind+1))

# create a new dataframe
neighborhoods_props_sorted = pd.DataFrame(columns=columns)
neighborhoods_props_sorted['GEO_ID'] = ac_grouped['GEO_ID']

for ind in np.arange(ac_grouped.shape[0]):
    neighborhoods_props_sorted.iloc[ind, 1:] = return_most_common_props(ac_grouped.iloc[ind, :], num_top_props)

    
#neighborhoods_venues_sorted.rename(index=str, columns={"GEOID": "GEO_ID"})   
neighborhoods_props_sorted.head(50)
```

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans


# set number of clusters
kclusters = 5

ac_grouped_clustering = ac_grouped.drop('GEO_ID', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(ac_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_props_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```
